```python
import pandas as pd

# Path to the file
file_path = "C:/Users/Admin/Downloads/AirQuality.csv"

# Load the dataset using the semicolon delimiter
data = pd.read_csv(file_path, delimiter=";", engine="python")

# Display the first few rows to check if the data loads correctly
print(data.head())
```

```python
# Check for missing values
print(data.isnull().sum())

# Option 1: Drop rows with missing values
data_cleaned = data.dropna()

# Convert columns to numeric values, forcing errors to NaN (i.e., non-numeric data will become NaN)
numeric_columns = data_cleaned.select_dtypes(include=['object']).columns
for col in numeric_columns:
    data_cleaned[col] = pd.to_numeric(data_cleaned[col], errors='coerce')
    
# Check data types after conversion
print(data_cleaned.dtypes)

```

```python
# Drop columns with a high number of missing values (e.g., 'Unnamed: 15' and 'Unnamed: 16')
data_cleaned = data_cleaned.drop(columns=['Unnamed: 15', 'Unnamed: 16'])

# Save the cleaned data to a new CSV file
data_cleaned.to_csv("C:/Users/Admin/Downloads/AirQuality_Cleaned_Final.csv", index=False)

# Display the first few rows to confirm the changes
print(data_cleaned.head())
```

```python
# Check the shape and a few rows of the original data
print(data.shape)
print(data.head())
```

```python
# Clean Date and Time columns to a datetime format
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')
data['Time'] = pd.to_datetime(data['Time'], format='%H.%M.%S').dt.time

# Drop the 'Unnamed' columns
data_cleaned = data.drop(columns=['Unnamed: 15', 'Unnamed: 16'])

# Print the cleaned data
print(data_cleaned.head())
```

```python
# Replace commas with dots for numeric columns
cols_to_clean = ['CO(GT)', 'T', 'RH', 'AH']
data_cleaned[cols_to_clean] = data_cleaned[cols_to_clean].replace({',': '.'}, regex=True)

# Convert columns to numeric type
data_cleaned[cols_to_clean] = data_cleaned[cols_to_clean].apply(pd.to_numeric, errors='coerce')

# Print the cleaned data
print(data_cleaned.head())
```

```python
# Check for missing or invalid 'Date' and 'Time' values
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')  # This will turn invalid dates into NaT
data['Time'] = data['Time'].astype(str)

# Drop rows where 'Date' or 'Time' is NaT (missing values)
data_cleaned = data.dropna(subset=['Date', 'Time'])

# Combine 'Date' and 'Time' into a single Datetime column
data_cleaned['Datetime'] = pd.to_datetime(data_cleaned['Date'].astype(str) + ' ' + data_cleaned['Time'], format='%Y-%m-%d %H:%M:%S')

# Plot time series for CO(GT) and NOx(GT)
plt.figure(figsize=(14, 6))
plt.plot(data_cleaned['Datetime'], data_cleaned['CO(GT)'], label='CO(GT)', color='tab:blue')
plt.plot(data_cleaned['Datetime'], data_cleaned['NOx(GT)'], label='NOx(GT)', color='tab:orange')
plt.xlabel('Time')
plt.ylabel('Concentration')
plt.title('Time Series of CO(GT) and NOx(GT)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Adjusting plot style globally
plt.rcParams['axes.grid'] = True  # Add grid to all plots
plt.rcParams['xtick.direction'] = 'inout'  # Make x-axis ticks inward and outward
plt.rcParams['ytick.direction'] = 'inout'  # Make y-axis ticks inward and outward

# Define a function to adjust tick labels and layout for all plots
def adjust_ticks_and_layout():
    plt.xticks(rotation=45, fontsize=10)  # Rotate x-axis labels and adjust font size
    plt.yticks(fontsize=10)  # Adjust y-axis font size
    plt.tight_layout(pad=3.0)  # Adjust layout to avoid overlap

# Example for plotting time series for multiple gases (CO, NMHC, NOx)
plt.figure(figsize=(14, 6))
plt.plot(data_cleaned['Datetime'], data_cleaned['CO(GT)'], label='CO(GT)', color='tab:blue')
plt.plot(data_cleaned['Datetime'], data_cleaned['NMHC(GT)'], label='NMHC(GT)', color='tab:green')
plt.plot(data_cleaned['Datetime'], data_cleaned['NOx(GT)'], label='NOx(GT)', color='tab:orange')
plt.xlabel('Time')
plt.ylabel('Concentration')
plt.title('Time Series of CO(GT), NMHC(GT), and NOx(GT)')
plt.legend()
adjust_ticks_and_layout()  # Apply global adjustments
plt.show()

# Example for another plot (e.g., for NO2(GT) and Ozone concentrations)
plt.figure(figsize=(14, 6))
plt.plot(data_cleaned['Datetime'], data_cleaned['NO2(GT)'], label='NO2(GT)', color='tab:red')
plt.plot(data_cleaned['Datetime'], data_cleaned['PT08.S4(NO2)'], label='PT08.S4(NO2)', color='tab:purple')
plt.plot(data_cleaned['Datetime'], data_cleaned['PT08.S5(O3)'], label='PT08.S5(O3)', color='tab:cyan')
plt.xlabel('Time')
plt.ylabel('Concentration')
plt.title('Time Series of NO2(GT), PT08.S4(NO2), and Ozone')
plt.legend()
adjust_ticks_and_layout()  # Apply global adjustments
plt.show()

# Additional visualizations can follow the same pattern
```
